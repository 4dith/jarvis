```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create a figure
fig, ax = plt.subplots(figsize=(10, 7))

# Turn off axis
ax.axis("off")

# Draw main brain (Intent Recognizer + Skill Manager)
brain = mpatches.FancyBboxPatch((0.35, 0.65), 0.3, 0.15,
                                boxstyle="round,pad=0.02", 
                                edgecolor="black", facecolor="#cce5ff")
ax.add_patch(brain)
ax.text(0.5, 0.725, "Core JARVIS\n(Intent Recognizer + Skill Manager)", 
        ha="center", va="center", fontsize=10, weight="bold")

# Interfaces (Top Row)
interfaces = [
    ("Website Chatbot", 0.1, 0.9),
    ("Discord Bot", 0.35, 0.9),
    ("Android/PC App", 0.6, 0.9),
    ("CLI", 0.85, 0.9)
]
for name, x, y in interfaces:
    rect = mpatches.FancyBboxPatch((x-0.1, y-0.05), 0.2, 0.08, 
                                   boxstyle="round,pad=0.02",
                                   edgecolor="black", facecolor="#ffe6cc")
    ax.add_patch(rect)
    ax.text(x, y-0.01, name, ha="center", va="center", fontsize=9)
    ax.annotate("", xy=(x, y-0.05), xytext=(0.5, 0.65),
                arrowprops=dict(arrowstyle="->", lw=1.5))

# Skills (Bottom Row)
skills = [
    ("FAQ Skill\n(Vector DB)", 0.15, 0.35),
    ("Event Schedule\n(Google Calendar / Notion)", 0.35, 0.35),
    ("Directory Skill\n(Member/Projects DB)", 0.55, 0.35),
    ("Device Control\n(MQTT/ROS)", 0.75, 0.35),
    ("General Knowledge\n(Web Search + Summarize)", 0.5, 0.15)
]

for name, x, y in skills:
    rect = mpatches.FancyBboxPatch((x-0.12, y-0.05), 0.24, 0.1, 
                                   boxstyle="round,pad=0.02",
                                   edgecolor="black", facecolor="#d5f5e3")
    ax.add_patch(rect)
    ax.text(x, y, name, ha="center", va="center", fontsize=8)
    ax.annotate("", xy=(x, y+0.05), xytext=(0.5, 0.65),
                arrowprops=dict(arrowstyle="->", lw=1.5))

plt.show()
```
